```python
labels=['a','b','c']
values=[1,2,3]
bars = plt.bar(labels,values)
#条形图的填充物
# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')
patterns=['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop())
#改变条形图的大小
plt.figure(figsize=(6,4))#这里没有指定dpi，那就是默认的dpi
plt.show()
```

```python
fifa=pd.read_csv('matplotlib_tutorial-master/fifa_data.csv')
fifa.head(5)
```

```python
plt.hist(fifa.Overall,bins=[40,50,60,70,80,90,100],color='b')
plt.xlabel("能力")
plt.ylabel("球员数量")
plt.title('fifa球员概况',fontdict={"fontsize":28})
plt.show()
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
plt.figure(figsize=(6,2),dpi=200)
plt.pie([left,right],labels=['left','right'],colors=['b','g'],autopct='%.2f%%')
plt.title('球员习惯使用左脚和右脚的比例')
plt.show()
```

```python
plt.figure(figsize=(3,5),dpi=300)
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight>=125)&(fifa.Weight<150)].count()[0]
medium = fifa[(fifa.Weight>=150)&(fifa.Weight<175)].count()[0]
medium_heavy = fifa[(fifa.Weight>=175)&(fifa.Weight<200)].count()[0]
heavy = fifa[(fifa.Weight>=200)].count()[0]
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['轻于125','125~150之间','150~175之间','175~200之间','比200还重']
explode = (0.4,0.2,0,0,0.4)
plt.pie(weights,labels=labels,autopct='%.2f%%',pctdistance=0.82,explode=explode)
plt.show()
```

```python
plt.figure(figsize=(5,8),dpi=300)
barcelona=fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club=='Real Madrid']['Overall']
revs = fifa.loc[fifa.Club=='New England Revolution']['Overall']
labels = ['FC Barcelona','Real Madrid','New England Revolution']
boxes = plt.boxplot([barcelona,madrid,revs],labels=labels,patch_artist=True,medianprops={'linewidth':2})
for box in boxes['boxes']:
    box.set(color='#213482',linewidth=2)
    box.set(facecolor='#324583')
plt.title('球队对比')
plt.ylabel('分值')
plt.show()
```
