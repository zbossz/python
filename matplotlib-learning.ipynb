```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from matplotlib import matplotlib_fname
matplotlib_fname()
```

```python
# resize my graph
plt.figure(figsize=(4,4),dpi=100)
plt.plot([1,2,3],[4,5,6],label='i dont know',color='green',linewidth='4',marker='*',markersize=10,markeredgecolor='red',linestyle='--')
plt.plot([1.2,2.3,4.5],[4,6.7,7.8],'r^--',label='what')#colormarkerline
plt.title("learning matplotlib",fontdict={"fontname":"Microsoft JhengHei","fontsize":20})
plt.xlabel("x轴",fontdict={"fontname":"Microsoft JhengHei","fontsize":16})
plt.ylabel("y轴",fontdict={"fontname":"Microsoft JhengHei","fontsize":16})
plt.yticks([0,1,2,3,4,5,6])
plt.xticks([0,1,2,3,4,5])#手动标刻度
plt.legend()
# save my graph
plt.savefig('matplotlib-graph1.jpg',dpi=300)
plt.show()
```
