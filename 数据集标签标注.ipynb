```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

# 乳腺癌

```python
cancer_data = pd.read_csv('./ML_dataset/breast cancer_dataset.csv')
cancer_data.head()
```

m晚期
b早期

```python
cancer_data['diagnosis'].value_counts()
```

```python
label_encode = LabelEncoder()
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
labels
```

```python
cancer_data['diagnosis']=labels
cancer_data.head()
```

为什么1是m，0是b。在我们没有指定的情况下，按字母顺序排，b在m前面，所以b是0，m是1.
